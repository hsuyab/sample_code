```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np
import matplotlib.pyplot as plt

df_clean = df.dropna(subset=['lat', 'lon'])
coords = df_clean[['lat', 'lon']].values

# Silhouette Score for different eps values (DBSCAN)
eps_values = np.arange(0.005, 0.1, 0.005)
silhouette_scores_dbs = []

for eps in eps_values:
    clustering_dbs = DBSCAN(eps=eps, min_samples=2).fit(coords)
    if len(set(clustering_dbs.labels_)) > 1:  # Check if more than one cluster is formed
        sil_score_dbs = silhouette_score(coords, clustering_dbs.labels_)
    else:
        sil_score_dbs = -1  # Set a negative score if only one cluster is formed
    silhouette_scores_dbs.append(sil_score_dbs)

optimal_eps = eps_values[np.argmax(silhouette_scores_dbs)]
print(f"Optimal eps for DBSCAN: {optimal_eps}")

# Davies-Bouldin Index for different eps values (DBSCAN)
db_scores_dbs = []

for eps in eps_values:
    clustering_dbs = DBSCAN(eps=eps, min_samples=2).fit(coords)
    if len(set(clustering_dbs.labels_)) > 1:  # Check if more than one cluster is formed
        db_score_dbs = davies_bouldin_score(coords, clustering_dbs.labels_)
    else:
        db_score_dbs = np.inf  # Set a large value if only one cluster is formed
    db_scores_dbs.append(db_score_dbs)

optimal_eps_db = eps_values[np.argmin(db_scores_dbs)]
print(f"Optimal eps for DBSCAN (Davies-Bouldin Index): {optimal_eps_db}")

# Plotting Silhouette Score and Davies-Bouldin Index for DBSCAN
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(eps_values, silhouette_scores_dbs, marker='o')
ax1.set_xlabel('eps')
ax1.set_ylabel('Silhouette Score')
ax1.set_title('Silhouette Score vs. eps (DBSCAN)')

ax2.plot(eps_values, db_scores_dbs, marker='o')
ax2.set_xlabel('eps')
ax2.set_ylabel('Davies-Bouldin Index')
ax2.set_title('Davies-Bouldin Index vs. eps (DBSCAN)')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

df_clean = df.dropna(subset=['lat', 'lon'])
coords = df_clean[['lat', 'lon']].values

# Elbow Plot for KMeans
max_clusters = 40
inertia_values = []

for n_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(coords)
    inertia_values.append(kmeans.inertia_)

# Plotting Elbow Curve
plt.figure(figsize=(10, 5))
plt.plot(range(1, max_clusters + 1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve (KMeans)')
plt.show()

# Silhouette Score for different number of clusters (KMeans)
silhouette_scores_km = []

for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(coords)
    silhouette_scores_km.append(silhouette_score(coords, kmeans.labels_))

optimal_clusters_km = np.argmax(silhouette_scores_km) + 2
print(f"Optimal number of clusters for KMeans: {optimal_clusters_km}")

# Plotting Silhouette Score for KMeans
plt.figure(figsize=(10, 5))
plt.plot(range(2, max_clusters + 1), silhouette_scores_km, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters (KMeans)')
plt.show()
```

```python
#create a list of unique features in df['leasures_available'].unique()
leasures = df['leasures_available'].unique()
leasures = leasures[~pd.isnull(leasures)]
leasures = [item.split(',') for item in leasures]
leasures = [item.strip() for sublist in leasures for item in sublist]
leasures = list(set(leasures))
print(leasures)
print(len(leasures))
```

```python
#find the unique values in [i.lower() for i in leasures]
leasures = [i.lower() for i in leasures]
leasures = list(set(leasures))
print(leasures)
```

```python
# Function to check the presence of each leisure in 'leasures_available'
def check_leisure(row, leisure):
    if pd.isnull(row['leasures_available']):
        return 0
    # Split the string by comma, strip each item, and check if the leisure is in the list
    return int(leisure in [item.strip().lower() for item in row['leasures_available'].split(',')])

# Create new columns for each unique leisure
for leisure in leasures:
    df[f'leisure_{leisure}'] = df.apply(check_leisure, leisure=leisure, axis=1)

# Now 'df' has new binary columns indicating the presence of each leisure facility
# print(df.head())
```

```python
# Loading the dataset to perform the heatmap analysis as described
# file_path = '/mnt/data/train.csv'
data = pd.read_csv("train.csv")

# Calculate the price per area ratio
data['price_per_area'] = data['price'] / data['area']

# Bin latitude and longitude and compute the average price per area for each bin
lat_bins = np.linspace(data['lat'].min(), data['lat'].max(), 10)
lon_bins = np.linspace(data['lon'].min(), data['lon'].max(), 10)

data['lat_bin'] = pd.cut(data['lat'], bins=lat_bins, labels=False, include_lowest=True)
data['lon_bin'] = pd.cut(data['lon'], bins=lon_bins, labels=False, include_lowest=True)

grouped = data.groupby(['lat_bin', 'lon_bin'])['price_per_area'].mean().reset_index()
# pivot_table = grouped.pivot('lat_bin', 'lon_bin', 'price_per_area')  # Specify the values parameter
pivot_table = grouped.pivot(index='lat_bin', columns='lon_bin', values='price_per_area')


# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(pivot_table, origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='Average Price per Area')
plt.title('Heatmap of Average Price per Area by Latitude and Longitude')
plt.xlabel('Longitude Bin')
plt.ylabel('Latitude Bin')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Calculate the price per area ratio again, if not already present
data['price_per_area'] = data['price'] / data['area']

# Remove any rows with NaN values in 'lat', 'lon', or 'price_per_area' to avoid issues in the plot
data_clean = data.dropna(subset=['lat', 'lon', 'price_per_area'])

# Prepare data for the density plot
x = data_clean['lon']
y = data_clean['lat']
z = data_clean['price_per_area']

# Calculate the point density
xy = np.vstack([x, y])
z_density = gaussian_kde(xy)(xy)


# Sort the points by density, so that the densest points are plotted last
idx = z_density.argsort()
x, y, z_density = pd.DataFrame(x).iloc[idx], pd.DataFrame(y).iloc[idx], pd.DataFrame(z_density).iloc[idx]

fig, ax = plt.subplots(figsize=(12, 10))
scatter = ax.scatter(x, y, c=z_density, s=50, cmap='viridis', vmin=z_density.min(), vmax=z_density.max())
fig.colorbar(scatter, ax=ax, label='Density')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Heatmap of Price per Area Density by Latitude and Longitude')
plt.show()
```
